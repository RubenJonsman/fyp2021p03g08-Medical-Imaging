# Plotting every RGB combination

```python
ls = np.array([[[0]*3]*3953]*3953)
ls.shape
r = 0
g = 0
b = 0

for i in range(3953):
    for j in range(3953):
        ls[i][j][0] = r
        ls[i][j][1] = g
        ls[i][j][2] = b
        
        if r < 250:
            r+=1
            
        elif r == 250:
            r = 0
            g += 1
        
        if g == 250:
            g = 0
            b += 1
        
        if b == 250:
            b = 0
```

```python
print(ls.shape)
plt.imshow(ls)
```
