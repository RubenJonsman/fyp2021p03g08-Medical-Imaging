```python
def height_of_lesion_points(p1, p2):
    x_mask = list(mask.any(axis=0))
    y_mask = list(mask.any(axis=1))
    xlim = x_mask.index(p1), len(x_mask)-x_mask[::-1].index(p1)
    ylim = y_mask.index(p2), len(y_mask)-y_mask[::-1].index(p2)

    plt.imshow(mask, cmap="gray")
    for xval in xlim:
        plt.vlines(xlim, 0, len(y_mask), colors="r")
    for yval in ylim:
        plt.hlines(ylim, 0, len(x_mask), colors="r")
    plt.plot(xlim, ylim, c="lime")
    print("distance is = {}".format(((xlim[0]-ylim[0])**2+(xlim[1]-ylim[1])**2)**0.5))

height_of_lesion_points(1,1)
```

```python
asd = mask[ylim[0]:ylim[1], xlim[0]:xlim[1]]
height, width = asd.shape
print("height:", height,"\nwidth:", width)
plt.imshow(asd,cmap="gray")
```

```python
def measure_area_perimeter_2(mask, size):
    # Measure area: the sum of all white pixels in the mask image
    area = np.sum(mask)

    # Measure perimeter: first find which pixels belong to the perimeter.
    struct_el = morphology.disk(size)
    mask_eroded = morphology.binary_erosion(mask, struct_el)
    image_perimeter = mask - mask_eroded

    # Now we have the perimeter image, the sum of all white pixels in it
    perimeter = np.sum(image_perimeter)

    return area, perimeter
```

## Asymmetry

```python
def fold_vertically(mask):
    half_mask = mask[:,:len(mask)//2]
    reversed_half_mask = half_mask[:, ::-1]
    new_circle = np.concatenate((half_mask, reversed_half_mask),axis=1)
    return new_circle

def fold_horizontally(mask):
    half_mask = mask[:len(mask)//2, :]
    reversed_half_mask = half_mask[::-1, :]
    new_circle = np.concatenate((half_mask, reversed_half_mask),axis=0)
    return new_circle
```

```python
print(tuple(np.subtract(measure_area_perimeter(mask), 
                        measure_area_perimeter(fold_horizontally(mask)))))

print(tuple(np.subtract(measure_area_perimeter(mask),
                        measure_area_perimeter(fold_vertically(mask)))))
```

Here we see the difference between the original lesion and the symmetric version of the same lesion is big, meaning that the lesion is asymmetric.

```python
area_mask, peri_mask = measure_area_perimeter(mask)
area_diff, peri_diff = tuple(np.subtract(measure_area_perimeter(mask), measure_area_perimeter(fold_horizontally(mask))))

(area_diff/area_mask)*100
```

```python
# implement rotation at 30°, 60° and 90°
def asymmetry(mask):
    
    def fold_vertically(mask):
        half_mask = mask[:,:len(mask)//2]
        reversed_half_mask = half_mask[:, ::-1]
        new_circle = np.concatenate((half_mask, reversed_half_mask),axis=1)
        return new_circle

    def fold_horizontally(mask):
        half_mask = mask[:len(mask)//2, :]
        reversed_half_mask = half_mask[::-1, :]
        new_circle = np.concatenate((half_mask, reversed_half_mask),axis=0)
        return new_circle
    
    empty_lst = []
    
    # adding non rotated areas.
    area = measure_area_perimeter(mask)[0]
    horizontal = area-measure_area_perimeter(fold_horizontally(mask))[0]
    vertical = area-measure_area_perimeter(fold_vertically(mask))[0]
    empty_lst.append(horizontal)
    empty_lst.append(vertical)
    
    # adding the rotated areas.
    rot = [30,60,90]
    for i in rot: 
        mask = transform.rotate(mask, i)
        horizontal = area-measure_area_perimeter(fold_horizontally(mask))[0]
        vertical = area-measure_area_perimeter(fold_vertically(mask))[0]
        empty_lst.append(horizontal)
        empty_lst.append(vertical)
    
    # taking a mean of all these rotated areas.
    mean = sum(empty_lst)/len(empty_lst)
    return mean
```

## Classifiers 3.1

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
train_scores = {}
test_scores = {}
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[["area","perimeter", "compactness"]],
                                                    df["diagosis_sum"], test_size=0.2, random_state=42)
```

```python
scores={}
scores_list=[]
for k in range(1,26):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred=knn.predict(X_test)
    scores[k] = accuracy_score(y_test, y_pred)
    scores_list.append(accuracy_score(y_test, y_pred))
plt.plot(range(1,26), scores_list);
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
y_pred=knn.predict(X_test)
score = accuracy_score(y_test, y_pred)
score
```

```python
DTC = DecisionTreeClassifier(random_state=42)
DTC.fit(X_train, y_train)
y_pred=DTC.predict(X_test)
test_scores["DecisionTree"] = accuracy_score(y_test, y_pred)
train_scores["DecisionTree"] = accuracy_score(y_train, DTC.predict(X_train))
```

```python
scores={}
scores_list=[]
for k in range(1,26):
    RFC = RandomForestClassifier(n_estimators=100, max_depth=k, random_state=42)
    RFC.fit(X_train, y_train)
    y_pred=RFC.predict(X_test)
    scores[k] = accuracy_score(y_test, y_pred)
    scores_list.append(accuracy_score(y_test, y_pred))
plt.plot(range(1,26), scores_list);
```

```python
RFC = RandomForestClassifier(max_depth=1, random_state=42)
RFC.fit(X_train, y_train)
y_pred=RFC.predict(X_test)
test_scores["RandomForest"] = accuracy_score(y_test, y_pred)
train_scores["RandomForest"] = accuracy_score(y_train, RFC.predict(X_train))
```

```python
SV = SVC() 
SV.fit(X_train, y_train)
y_pred=SV.predict(X_test)
test_scores["SupportVector"] = accuracy_score(y_test, y_pred)
train_scores["SupportVector"] = accuracy_score(y_train, SV.predict(X_train))
```

```python
LR = LogisticRegression()
LR.fit(X_train, y_train)
y_pred=LR.predict(X_test)
test_scores["LogisticRegression"] = accuracy_score(y_test, y_pred)
train_scores["LogisticRegression"] = accuracy_score(y_train, LR.predict(X_train))
```
